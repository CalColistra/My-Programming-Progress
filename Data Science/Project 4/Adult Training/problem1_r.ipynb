# **Use the following dataset about adults:**

```r
#read in the data file:
library(readr)
adults <- read_csv("https://raw.githubusercontent.com/CalColistra/IDS/main/Project4/problem1/adult_training.csv", show_col_types = FALSE)
adults
```

# **Create a contingency table of workclass and sex, and plot it.**

```r
#Cross tabulation of workclass and sex:
r_crosstab <- table(adults$sex, adults$workclass)
ftable(r_crosstab)
colors <- c("orange", "blue")
#plot it:
barplot(r_crosstab,las = 2, cex.names = 1, col = colors, beside = TRUE)
legend("topright", legend = c("Female", "Male"), fill = colors)
```

# **Create a contingency table of sex and marital status, and plot it.**

```r
#Cross tabulation of sex and marital status:
r_crosstab <- table(adults$sex, adults$maritalStatus)
ftable(r_crosstab)
#plot it:
colors <- c("orange", "blue")
barplot(r_crosstab, las = 2, cex.names = 1, col = colors, beside = TRUE)
legend("topright", legend = c("Female", "Male"), fill = colors)
```

# **Create a new data frame, adultOver40, for those whose age > 40.**

```r
#make new dataFrame without rows where age is over 40:
adultOver40 <-adults[(adults$age > 40),]
#print the new df:
adultOver40
```

# **Recreate the contingency table of sex and marital status for adultOver40.**

```r
#Cross tabulation of sex and marital status:
r_crosstab2 <- table(adultOver40$sex, adultOver40$maritalStatus)
ftable(r_crosstab2)
#plot it:
colors <- c("orange", "blue")
barplot(r_crosstab2, las = 2, cex.names = 1, col = colors, beside = TRUE)
legend("topright", legend = c("Female", "Male"), fill = colors)
```

# **Determine whether any outliers exist for the education field.**

```r
#make a new column and put the scaled education into it:
adults$scaled_education <- scale(x = adults$education)
#summarize & make sure mean = 0:
summary(adults$scaled_education)
```

```r
#access outliers:
three_sigma_rule <- (adults$scaled_education < -3 | 
                     adults$scaled_education > 3 )
#print how many outliers are there:
cat ("There are", sum(three_sigma_rule), "outliers in the education field \n\n")
```
